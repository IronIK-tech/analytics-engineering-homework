```python
import pandas as pd
import numpy as np

data = {
    'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
    'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
    'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']
}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
```

```python
# 1. Створити DataFrame
df = pd.DataFrame(data)
```

```python
# 2. Змінити індекси
df.index = labels
```

```python
# 3. Значення в колонці 'animal', рядок з індексом 'e'
df.loc['e', 'animal']
```

```python
# 4. Кількість непорожніх значень у колонці 'age'
df['age'].count()
```

```python
# 5. Перші 3 рядки таблиці
df.head(3)
```

```python
# 6. Рядки з позиціями 0, 2, 3
df.iloc[[0, 2, 3]]
```

```python
# 7. Тільки стовпці 'animal' та 'age'
df[['animal', 'age']]
```

```python
# 8. Стовпці 'animal' та 'age', рядки з позиціями 0, 2, 3
df.iloc[[0, 2, 3]][['animal', 'age']]
```

```python
# 9. Рядки, де age > 3
df[df['age'] > 3]
```

```python
# 10. Рядки, де age не є порожнім
df[df['age'].notna()]
```

```python
# 11. Рядки, де age від 2 до 4 включно
df[df['age'].between(2, 4)]
```

```python
# 12. Збільшити age на 1 для всіх тварин
df['age'] = df['age'] + 1

df
```

```python
# 13. Рядок з максимальним значенням age
df.loc[df['age'] == df['age'].max()]
```

```python
# 14. Сума всіх значень у колонці 'age'
df.loc[:, 'age'].sum()
```

```python
# 15. Рядки, де animal == 'cat'
df.loc[df['animal'] == 'cat']
```

```python
# 16. Рядки, де animal == 'dog' і visits > 1
df.loc[(df['animal'] == 'dog') & (df['visits'] > 1)]
```

```python
# 17. Рядки, де priority == 'yes'
df.loc[df['priority'] == 'yes']
```

```python
# 18. Рядки, де priority == 'no' і age >= 3
df.loc[(df['priority'] == 'no') & (df['age'] >= 3)]
```

```python
# 19. Колонки 'animal' та 'visits' де visits == 1
df.loc[df['visits'] == 1, ['animal', 'visits']]
```

```python
# 20. Рядки, де animal == 'cat' або 'snake'
df.loc[df['animal'].isin(['cat', 'snake'])]
```

```python
# 21. Рядки, де age < 3 або age є порожнім
df.loc[(df['age'] < 3) | (df['age'].isna())]
```

```python
# 22. Рядки, де visits від 2 до 3 включно
df.loc[df['visits'].between(2, 3)]
```

```python
# 23. Рядки де індекс == 'a', 'c' або 'f'
df.loc[['a', 'c', 'f']]
```

```python
# 24. Рядки де animal != 'dog'
df.loc[df['animal'] != 'dog']
```
